# Selenium에서 주로 활용하는 함수 

1. get : URL에 접속하게 해주는 함수
    * <font color = "green">ex )  div.get('https://www.naver.com/') 


2. find_element_by_xpath : 해당 홈페이지에서 가져온 **Xpath**를 활용하여 원하는 값을 가져오게 해주는 함수
    * <font color = "green">ex )  div.find_element_by_xpath('//*[@id="query"]')
    

3. find_element_by_css_selector : 홈페이지에서 가져온 **Copy selector**를 활용하여 원하는 값을 가져오게 해주는 함수 
    * <font color = "green">ex )  div.find_elemen_by_css_selector('#account > div > a > i')
    

4. click() : xpath나 css_selector을 활용하여 불러온 위치를 클릭하게 해주는 함수
    * <font color = "green">ex )  div.find_element_by_xpath('//*[@id="search_btn"]').click()
    

5. send_keys() : xpath나 css_selector을 활용하여 불러온 위치에 원하는 값을 입력해준다.  
    * <font color = "green">ex )  div.find_element_by_xpath('//*[@id="query"]').send_keys('우한 폐렴')
    

6. clear() : xpath나 css_selector을 활용하여 불러온 위치에 입력한 값을 모두 지워준다. 
    * <font color = "green">ex )  div.find_element_by_xpath('//*[@id="query"]').clear()


7. time.sleep() : ()안에 숫자를 입력하면 해당 시간(초)만큼 잠시 대기. 주로 셀레니움에서 홈페이지 로딩시간이나 스크롤을 내릴때 대기시간과 같이 페이지의 로딩이 필요할때나 데이터를 끌어올 때 서버에서 차단당하는 일을 최대한 막기위해 주로 사용한다.
    * <font color = "green">ex )  time.sleep(3)

# 크롤링할 때 xpath 규칙 알아보기
1. 가져오고 데이터를 찾는다. 


2. 해당 위치의 xpath를 가져온다.


3. 다음 위치에 있는 xpath를 가져온다.


4. url처럼 비교 후 규칙을 찾아본다.


5. 실행한 후 결과가 나온다면 해당 방식을 활용하여 반복문 등을 이용 내가 원하는 횟수만큼 가져올 수 있게 한다. 

```python
//*[@id="old_content"]/table/tbody/tr[2]/td[2]/div/a # 1위 검색어 xpath
//*[@id="old_content"]/table/tbody/tr[3]/td[2]/div/a # 2위 검색어 xpath
...
//*[@id="old_content"]/table/tbody/tr[11]/td[2]/div/a # 10위 검색어 xpath
# tr을 기준으로 랭킹 순위가 바뀐다는 것을 알 수 있다.
```

```python
# tr위치에 값을 바꾸어 가면서 xpath 지정
# 전체연령대 1등부터 20등까지 가져오기 위해 
xpath = []
for i in range(2,12):
    xpath.append('//*[@id="old_content"]/table/tbody/tr[' + str(i) + ']/td[2]/div/a')
xpath
```

## URL과 xpath를 활용하여 1위 부터 50위 까지 영화명 가져오기 - for문

### 과정
1. 순위를 가진 DataFrame을 만든다.


2. 바깥에 있는 반복문을 활용하여 URL에 접속하게 만든다. 


3. 빈 리스트를 만들어 해당 리스트에 영화명가 들어갈 수 있도록 한다.


4. URL 반복문 내에 xpath 반복문을 활용하여 1위부터 10위까지의 급상승 검색어를 가져와서 빈 리스트에 저장한다.


5. 리스트가 채워지면 해당 리스트를 미리 만들어둔 데이터 프레임에 넣는다. 이때 제목은 위에서 url을 만들때 지정해 두었던 날짜를 집어 넣는다. 


6. 만들어진 데이터를 확인한다.

```python
# 급상승 검색어 크롤링 (네이버 데이터랩 기준)
# 자세히 설명 
data = pd.DataFrame({'ranking':range(1,11)}) # 1등부터 20등까지

for i in range(len(url)):
    ranking = []
    div.get(url[i]) # 접속 url
    time.sleep(2)
    
    for j in range(10):
        ranking.append(div.find_element_by_xpath(xpath[j]).text) # xpath이용 1위부터 검색어 키워드 가져옴. 모든 연령대
    
    data[date[i]] = ranking
    
div.close() 

data
```

## from Wide form to Long form

```python
data = pd.melt(data,id_vars='ranking')
data
```

```python
data['month'] = data.variable.dt.month # 월뽑기
data['day'] = data.variable.dt.day # 일뽑기
data['dayOfweek'] = data.variable.dt.dayofweek #요일뽑기
data['week'] = data.variable.dt.week # 주뽑기
data
```

# 과제 ) 영화 랭킹 상위 10위 크롤링하는 함수만들기 

```python
# 일자를 입력하면 해당일자의 상위 10위 영화 데이터 프레임 반환
# date는 '20200201' 형태로 받음.
# 예시) TOP10_Movie('20200201')

def TOP10_Movie(date):
    div = wd.Chrome('C:/Users/LG/Downloads/chromedriver.exe') 
    
    # URLs list
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&tg=0&date=' 
    URLs = [url + d for d in date]
    
    # Xpath list
    xpath = []
    for i in range(2,12):
        xpath.append('//*[@id="old_content"]/table/tbody/tr[' + str(i) + ']/td[2]/div/a')
    
    data = pd.DataFrame({'ranking':range(1,11)}) # 1등부터 20등까지

    for i in range(len(URLs)):
        ranking = []
        div.get(URLs[i]) # 접속 url
        time.sleep(2)
    
        for j in range(10):
            ranking.append(div.find_element_by_xpath(xpath[j]).text) # xpath이용 1위부터 검색어 키워드 가져옴. 모든 연령대
    
        data[date[i]] = ranking
    
    div.close()
    
    data = pd.melt(data,id_vars='ranking')
    return data
```
